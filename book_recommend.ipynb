```python
##Popularity based recommender system
```

```python
rating_books=books_df.merge(rating_df,on="ISBN")
rating_books.shape
```

```python
num_rating_df=rating_books.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={"Book-Rating":"num_rating"},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_books.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
popularity_df
```

```python
popularity_df=popularity_df[popularity_df["num_rating"]>=250].sort_values("avg_rating",ascending=False).head(100)
```

```python
popularity_df
```

```python
popularity_df=popularity_df.merge(books_df,on="Book-Title").drop_duplicates("Book-Title")
```

```python
popularity_df
```

```python
#Collaborative Filtering Based Recommender System
```

```python
x=rating_books.groupby("User-ID").count()["Book-Rating"]>200
new_df=x[x].index
```

```python
filtered_rating=rating_books[rating_books["User-ID"].isin(new_df)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index= np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books_df[books_df['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
